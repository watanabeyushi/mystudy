**ステップ関数の実装**

```python
##step関数（実数のみを引数）
def step_function(x):
  if x>0:
    return 1
  else:
    return 0

##step関数（配列対応）
def step_function(x):
  y = x > 0
  return y.astype(int)
```

```python
import numpy as np
x=np.array([-1.0,1.0,2.0])
print(x)
y=x>0
print(y)
y=y.astype(int)
print(y)
```

**ステップ関数のグラフ**

```python
import numpy as np
import matplotlib.pylab as plt

def step_function(x):
  return np.array(x>0,dtype=int)

x=np.arange(-5.0,5.0,0.1)
y=step_function(x)
plt.plot(x,y)
plt.ylim(-0.1,1.1)
plt.show()            ##ｙ軸の範囲を指定
```

**シグモイド関数の実装**

```python
def sigmoid(x):
  return 1/(1+np.exp(-x))

x=np.array([-1.0,1.0,2.0])
sigmoid(x)
```

```python
t=np.array([1.0,2.0,3.0])
print(1.0+t)
print(1.0/t)
```

```python
x=np.arange(-5.0,5.0,0.1)
y=sigmoid(x)
plt.plot(x,y)
plt.ylim(-1.0,1.1)  ##ｙ軸の範囲を指定
plt.show()
```
