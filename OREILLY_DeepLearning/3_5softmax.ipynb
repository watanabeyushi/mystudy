**例**

```python
import numpy as np

a=np.array([0.3,2.9,4.0])

exp_a=np.exp(a)
print(exp_a)
sum_exp_a=np.sum(exp_a)
print(sum_exp_a)
y=exp_a/sum_exp_a
print(y)
```

**ソフトマックス**

```python
def softmax(a):
  exp_a = np.exp(a)
  sum_exp_a=np.sum(exp_a)
  y=exp_a/sum_exp_a
  return y
```

```python
##overflow
a=np.array([1010,1000,990])
print(np.exp(a)/np.sum(np.exp(a)))
```

**オーバーフロー対策したソフトマックス関数**

```python
def softmax(a):
  c=np.max(a)
  exp_a=np.exp(a-c)
  sum_exp_a=np.sum(exp_a)
  y=exp_a/sum_exp_a
  return y
```

```python
a=np.array([0.3,2.9,4.0])
y=softmax(a)
print(y)
print(np.sum(y))
```
